# Initialization

Initilization of restart handler class reads in the `id0` file of the restart dump. Key members are
* `par`: stores the input parameter block
* `par_misc`: stores non-data part in raw binary format
* `grids`: construct mpi grids

```python
rh = sim.load_rst(sim.nums_rst[10])
```

```python
data = rh.read()
```

# Data manipulation

Currently, restart handler support `refine`, `degrade`, and `cut_z` manipulations. See examples below. After manipulation, the manipulated data is stored to `data_target` as well as new `grids` and `par` are set and stored into corresponding the `grids_target` and `par_targer` members. `remove_all_targets` is a method to clean up all manipulated data and information.

# Progressively degrade data

Degrading manimulation takes volume and area weighted averages for 8 cells and 4 faces of cell and face centered variabels, respectively. After manipulation, it autamatically recalculate `grids` and `par` by calling `reset_grids` and `reset_par` methods. `reset_par` is called within `reset_grids` as `reset_par` must be called after `reset_grids`

```python
fig, axes = plt.subplots(1,5,figsize=(15,5))
plt.sca(axes[0])
plt.imshow(rh.data['DENSITY'].mean(axis=0),norm=LogNorm(vmin=1.e-1,vmax=1.e1))

# clean up any previous manipulated data
rh.remove_all_targets()

for i,ax in enumerate(axes[1:]):
    plt.sca(ax)
    rh.degrade();
    plt.imshow(rh.data_target['DENSITY'].mean(axis=0),norm=LogNorm(vmin=1.e-1,vmax=1.e1))
#     ax.axis('off')
```

# Cutting vertical domain

We can cut off extraplanar regions using the `cut_z` method

```python
fig, axes = plt.subplots(1,5,figsize=(15,10))
plt.sca(axes[0])
plt.imshow(rh.data['DENSITY'].mean(axis=1),norm=LogNorm(vmin=1.e-1,vmax=1.e1))

for i,ax in enumerate(axes[1:]):
    plt.sca(ax)
    zmin=rh.dm['x3min']+256*(i+1)
    zmax=rh.dm['x3max']-256*(i+1)
    rh.remove_all_targets() # clean up previously manipulated data
    rh.cut_z(zmin,zmax); # this autoamtically call reset_grids
    plt.imshow(rh.data_target['DENSITY'].mean(axis=1),norm=LogNorm(vmin=1.e-1,vmax=1.e1))
```

# Cut and refine
Finally, we refine the last data that has cut `|z|>768pc`. The `refine` method simple increase the number of dimensions by a factor of 2 copying cell/face centered variables into 8/4 nearby cells/faces.

```python
# we don't clean up targets here
rh.refine();
plt.imshow(rh.data_target['DENSITY'].mean(axis=1),norm=LogNorm(vmin=1.e-1,vmax=1.e1))
```

# write output

All manipulation has changed `data`, `par`, `grids`, `NGrid`. Simple write method will write new restart dumps for manipulated data and information (i.e. `data_target` not `data`). From the last manipulation, we have `target` for the  refined data with a vertical cut.

```python
# currently, we didn't change the grid size
print(rh.ngrid,rh.ngrid_target)
# Although the domain is reduced, we refined the data so that we now have more grids within the domain
print(rh.NGrid,rh.NGrid_target)
```

```python
# if we want to use larger grid, we can call reset_grids method again with a specified ngrid
rh.reset_grids(ngrid=[64,64,64])
print(rh.ngrid,rh.ngrid_target)
print(rh.NGrid,rh.NGrid_target)
```

```python
# finally, calling `write` method will generate 192 files for new data
new_rst_fname = rh.write(pid='LGR4_2pc_small')
```
